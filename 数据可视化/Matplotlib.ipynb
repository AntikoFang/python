```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

a = np.arange(0.0,5.0,0.02)

plt.subplot(211)
plt.plot(a,f(a))

plt.subplot(212)
plt.plot(a, np.cos(2*np.pi*a),'r--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplot2grid((3,3), # 设置一个 3*3 的网格
                 (0,0), # 选择左上角的一格
                 colspan = 3) # 横向扩展3列
plt.subplot2grid((3,3),
                 (1,0), # 选择第二行的第一格
                 colspan = 2) # 横向扩展两列
plt.subplot2grid((3,3),
                 (1,2), # 选择第二行的第三格
                 rowspan = 2) # 纵向扩展两行
plt.subplot2grid((3,3),(2,0))
plt.subplot2grid((3,3),(2,1))

plt.show()
```

```python
import matplotlib.gridspec as gridspec
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

a = np.arange(0.0,5.0,0.02)
b = np.arange(10)

gs = gridspec.GridSpec(3,3) #设置 3*3 的区域

ax1 = plt.subplot(gs[0,:])
plt.plot(a,f(a))

ax2 = plt.subplot(gs[1, :-1])
plt.plot(a, np.cos(2*np.pi*a), 'r--',*(a, np.sin(2*np.pi*a), 'b--'))

ax3 = plt.subplot(gs[1:, -1])
plt.plot(a, np.cos(2*np.pi*a),'r--')

ax4 = plt.subplot(gs[2, 0])
plt.plot(b, b*1.5, b, b*2.0, b, b*2.5)

ax5 = plt.subplot(gs[2, 1])
plt.plot(b, b*1.5,'go-', b, b*2.0,'rx',b, b*2.5,'*')

plt.show()
```

## pyplot的plot()函数

```python
plt.plot(x,y,format_string,**kwargs)
# x:x轴数据 列表或数组
# y:y轴数据 列表或数组
# format_string:控制曲线的格式字符串
# **kwargs:第二组或更多(x,y,format_string)
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(10)
plt.plot(a, a*1.5, a, a*2.0, a, a*2.5)#由三组x，y构成
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(10)
plt.plot(a, a*1.5,'go-', a, a*2.0,'rx',a, a*2.5,'*')#'go-':绿色实心圈实线；'rx':红色x标记；'*':系统自定义颜色星号标记
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='SimHei'#'SimHei'为黑体
matplotlib.rcParams['axes.unicode_minus'] = False #显示负号
matplotlib.rcParams['font.size']=20
plt.plot([9,3,1,0,1,5])
plt.ylabel("纵轴（值）")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams['font.family']='STSong'
matplotlib.rcParams['font.size']=20

a = np.arange(0.0, 5.0, 0.02)

plt.xlabel("横轴：时间")
plt.ylabel("纵轴：振幅")
plt.plot(a, np.cos(2*np.pi*a), 'r--',*(a, np.sin(2*np.pi*a), 'b--'))
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

a = np.arange(0.0, 5.0, 0.02)

plt.xlabel("横轴：时间", fontproperties = 'STSong', fontsize = 20)
plt.ylabel("纵轴：振幅", fontproperties = 'STsong', fontsize = 20)
plt.plot(a, np.cos(2*np.pi*a), 'r--',*(a, np.sin(2*np.pi*a), 'b--'))
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

a = np.arange(0.0, 5.0, 0.02)
plt.plot(a, np.cos(2*np.pi*a), 'b-')

plt.xlabel("横轴：时间", fontproperties = 'STSong', fontsize = 18,color = 'orange')
plt.ylabel("纵轴：振幅", fontproperties = 'STsong', fontsize = 18)
plt.title(r'余弦波实例 $y = cos(2\pi x)$',fontproperties = 'SimHei',fontsize = 25)
plt.annotate(r'$\mu = 100$',xy=(2,1), xytext=(3,1.5), #文本所在坐标位置
             arrowprops = dict(facecolor = 'red',
                               shrink = 0.1, # 箭头两侧缩进，离目标坐标（2,1）一定距离
                               width = 2))
plt.text(2,1,r'$\mu = 100$',fontsize = 15) # (2,1)横纵坐标


plt.axis([-1,6,-2,2])
plt.grid(True) # 加入网格
plt.show()
```
