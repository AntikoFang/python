# plt.plot()其他常用函数

- linewidth() #定义线条的宽度，可取任意实数

drawstyle:{'default','steps','steps-pre','steps-mid','steps-post'}

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0,10,10)
y = np.cos(x)

gs = gridspec.GridSpec(3,2) #设置 3*3 的区域

ax1 = plt.subplot(gs[0,:])
plt.plot(x, y, 'b-',linewidth = '5', drawstyle ='default', alpha = 0.5)

ax2 = plt.subplot(gs[1,0])
plt.plot(x, y, 'b-',linewidth = '5', drawstyle ='steps', alpha = 0.5)
ax3 = plt.subplot(gs[1,1])
plt.plot(x, y, 'b-',linewidth = '5', drawstyle ='steps-pre', alpha = 0.5)

ax4 = plt.subplot(gs[2, 0])
plt.plot(x, y, 'b-',linewidth = '5', drawstyle ='steps-mid', alpha = 0.5)

ax5 = plt.subplot(gs[2, 1])
plt.plot(x, y, 'b-',linewidth = '5', drawstyle ='steps-post', alpha = 0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']

x = np.linspace(0.5,10,1000)
y = np.cos(x)
plt.plot(x, y, 'b-',linewidth = '10', drawstyle = 'steps', alpha = 0.5)
plt.title('cos(x)')
plt.xlim(0,10)
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

x = np.linspace(0.05,10,100)
y = np.cos(x)
plt.plot(x, y, 'b-',linewidth = '10', alpha = 0.5)
plt.title('cos(x)')
plt.xlim(0, 9)
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.grid()
plt.axvline(3, c = 'g', lw = 5, ls = '--')
plt.axhspan(0.35,0.65,facecolor = 'yellow',alpha = 0.6 )
plt.legend(loc = 1)
plt.text(7.5,0.8,'y = cos(x)', weight = 'bold', color = 'r')
plt.show()
```

# plt.gca() 与 plt.gcf()
## plt.gca()获得当前坐标轴

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

x = np.linspace(0.05,10,100)
y = np.sin(x)

plt.plot(x,y)
ax = plt.gca()
ax.spines["right"].set_color("red")  #spine为框框的线
ax.spines["left"].set_color("g")
ax.spines["top"].set_color("orange")
ax.spines["bottom"].set_color("blue")

ax.spines["right"].set_color("none") #可以把颜色设置为none让线消失
ax.spines["top"].set_color("none")

ax.xaxis.set_ticks_position('bottom') #把x轴的刻度设置为bottom
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data',0))  #设置bottom对应到0点
ax.spines['left'].set_position(('data',4.65))
plt.show()
```

# plt.figure() 创建figure对象
- figure对象是后序绘图操作的总容器
- 如果不创建figure对象，直接调用接下来的plot()进行绘图，matplotlib会自动创建一个figure对象
- 事先创建figure对象可以对它做更精细的设定
## figsize 纸的大小

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.figure('np.sin(x)',           #名字
           figsize=(9,3),         #纸张大小
           dpi = 500,             # 像素
           facecolor = 'purple',  #纸张的颜色
           edgecolor = 'pink',    #
           frameon = True,      #是否绘制图形外框架
           clear = False,         #图形已存在，是否清除
          ) 
x = np.linspace(0.05,10,100)
y = np.sin(x)

plt.plot(x,y)
ax = plt.gca()
ax.spines["right"].set_color("red")  #spine为框框的线
ax.spines["left"].set_color("g")
ax.spines["top"].set_color("orange")
ax.spines["bottom"].set_color("blue")

ax.spines["right"].set_color("none") #可以把颜色设置为none让线消失
ax.spines["top"].set_color("none")

ax.xaxis.set_ticks_position('bottom') #把x轴的刻度设置为bottom
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data',0))  #设置bottom对应到0点
ax.spines['left'].set_position(('data',4.65))

plt.show()

#plt.savefig('位置')   #保存图片
```
